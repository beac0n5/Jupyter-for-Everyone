## changing dtypes

```python
vars_to_cat = ['Attack category', 'Protocol', 'Source IP', 'Destination IP']
df[vars_to_cat] = df[vars_to_cat].astype('category') # category takes less space
df.info()
```

# data preprocessing

## drop columns

```python
df.drop(columns = ['.'], inplace = True)
```

## process Time variable

```python
df[['start_time', 'end_time']] = df.Time.str.split('-', expand = True)
df.drop(columns = ['Time'], inplace = True)
df.head()
```

## creating timediff variable

```python
df['timediff_s'] = (df['end_time'].astype('int') - df['start_time'].astype('int'))
```

```python
df.head()
```

```python
df.dtypes
```

```python
to_datetime_vars = ['start_time', 'end_time']
df[to_datetime_vars] = df[to_datetime_vars].astype('int').apply(pd.to_datetime, unit = 's')
df.head()
```

```python
df.dtypes
```

```python
def check_char(char):
    char = char.strip().lower()
    if 'cisco' in char:
        return 'cisco'
    elif 'windows' in char:
        return 'windows'
    elif 'microsoft' in char:
        return 'microsoft'
    elif 'domino' in char:
        return 'domino'
    else:
        return np.nan
```

```python
df['attack_name'].apply(lambda x: check_char(x))
```

## merging dataframes

### merge

```python
df_ports = pd.read_csv('TCP-ports.csv')
```

```python
df_ports.head()
```

```python
df_ports.info()
```

```python
df_merged = df.merge(df_ports, left_on = 'destination_port', right_on = 'Port', how = 'left')
```

```python
df_merged.shape
```

```python
df_merged.head()
```

### join

```python
df.set_index('destination_port', inplace = True)
df_ports.set_index('Port', inplace = True)
```

```python
df.head()
```

```python
df_joined = df.join(df_ports, how = 'left')
df_joined.sample(10)
```

```python
df_joined.shape
```

### union

```python
cols_to_strip = ['attack_category', 'attack_subcategory', 'protocol', 'attack_name', 'attack_reference']
df[cols_to_strip] = df[cols_to_strip].apply(lambda x: x.str.strip())
```

```python
df_union1 = df.loc[df['attack_category'] == 'Reconnaissance', :]
df_union2 = df.loc[df['attack_category'] == 'Exploits', :]
df_union3 = df.loc[df['attack_category'] == 'Fuzzers', :]
```

```python
df_union1.shape, df_union2.shape, df_union3.shape
```

```python
df_union = pd.concat([df_union1, df_union2, df_union3], axis = 0)
df_union.shape
```

# summarising data

## pivot_table

```python
df.pivot_table(index = 'protocol',
               columns = 'attack_category',
               values = 'timediff_s',
               aggfunc = 'mean')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df_pivot1 = df.pivot_table(index = 'protocol',
               columns = 'attack_category',
               values = 'timediff_s',
               aggfunc = ['mean', 'count'])
df_pivot1.head(5)
```

```python
df_pivot2 = df_merged.pivot_table(index = 'protocol',
               columns = ['attack_category', 'Service'],
               values = 'timediff_s',
               aggfunc = 'sum',
                fill_value = 0)
df_pivot2.head(10)
```

### crosstab

```python
df_crosstab = pd.crosstab(index = df['protocol'],
               columns = df['attack_category'],
               values = df['timediff_s'],
               aggfunc = 'sum')
df_crosstab.head(5)
```

### groupby

```python
df_merged.groupby(by = 'protocol')['timediff_s'].mean()\
    .sort_values(ascending = False)\
    .head(10)
```

```python
df_merged.groupby(by = 'weekday').agg({'timediff_s':['sum', 'count', 'mean', 'median']})
```

```python
df_agg = df_merged.groupby(by = ['weekday', 'day', 'hour']).agg({'timediff_s':['sum', 'count', 'mean', 'median'],
                                                       'Service':['count']})
df_agg
```

```python
cols = [f'{x[0]}_{x[1]}' for x in df_agg.columns.tolist()]
df_agg.columns = cols
df_agg.head()
```

# styling dataframes

```python
df_agg.style.format({'timediff_s_mean':'{:.2f}',
                    'timediff_s_median':'{:.2f}'})\
    .bar(subset = 'timediff_s_sum', color = 'lightgreen')\
    .bar(subset = 'timediff_s_count', color = 'orange')\
    .background_gradient(subset = 'timediff_s_mean', cmap = 'Blues')\
    .background_gradient(subset = 'Service_count', cmap = 'brg')
```

# saving to disk
