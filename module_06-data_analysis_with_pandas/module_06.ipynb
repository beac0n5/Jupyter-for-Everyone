# data preprocessing

## drop columns

```python
df.drop(columns = ['.'], inplace = True)
```

## process Time variable

```python
df[['start_time', 'end_time']] = df.Time.str.split('-', expand = True)
df.drop(columns = ['Time'], inplace = True)
df.head()
```

## creating timediff variable

```python
df['timediff_s'] = (df['end_time'].astype('int') - df['start_time'].astype('int'))
```

```python
df.dtypes
```

```python
to_datetime_vars = ['start_time', 'end_time']
df[to_datetime_vars] = df[to_datetime_vars].astype('int').apply(pd.to_datetime, unit = 's')
df.head()
```

```python
def check_char(char):
    char = char.strip().lower()
    if 'cisco' in char:
        return 'cisco'
    elif 'windows' in char:
        return 'windows'
    elif 'microsoft' in char:
        return 'microsoft'
    elif 'domino' in char:
        return 'domino'
    else:
        return np.nan
```

```python
df_merged.groupby(by = 'protocol')['timediff_s'].mean()\
    .sort_values(ascending = False)\
    .head(10)
```

```python
df_agg = df_merged.groupby(by = ['weekday', 'day', 'hour']).agg({'timediff_s':['sum', 'count', 'mean', 'median'],
                                                       'Service':['count']})
df_agg
```

```python
df_agg.columns
```

```python
cols = [f'{x[0]}_{x[1]}' for x in df_agg.columns.tolist()]
df_agg.columns = cols
df_agg.head()
```

# styling dataframes

```python
df_agg.style.format({'timediff_s_mean':'{:.2f}',
                    'timediff_s_median':'{:.2f}'})\
    .bar(subset = 'timediff_s_sum', color = 'lightgreen')\
    .bar(subset = 'timediff_s_count', color = 'orange')\
    .background_gradient(subset = 'timediff_s_mean', cmap = 'Blues')\
    .background_gradient(subset = 'Service_count', cmap = 'brg')
```
