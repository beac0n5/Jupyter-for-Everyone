# vocabulary in info column

```python
pattern = r'\b[a-zA-Z]+\b'
def get_unique(s, counter = False):
    if type(s) == pd.Series:
        s = s.tolist()
    filtered_list = [x for x in s if x is not None]
    flattened_list = [item for sublist in filtered_list for item in sublist]
    if counter:
        counts = Counter(flattened_list)
        return counts
    else:
        unique_set = set(flattened_list)
        return list(unique_set)
```

## small ddf

```python
ds_small = ddf_small['Info'].str.findall(pattern)
s_small = ds_small.map_partitions(get_unique, True).compute()
s_small
```

```python
df_unique_counts = pd.DataFrame.from_records(s_small.tolist()).T
df_unique_counts
```

## whole ddf

```python
%%time
ds_whole = ddf['Info'].str.findall(pattern)
s_whole = ds_whole.map_partitions(get_unique, True).compute()
s_whole
```

```python
df_unique_counts_whole = pd.DataFrame.from_records(s_whole.tolist()).T
df_unique_counts_whole
```

```python
df_voc_all_parts = df_unique_counts_whole.sum(axis = 1).sort_values(ascending = False)
df_voc_all_parts.head(50)
```
